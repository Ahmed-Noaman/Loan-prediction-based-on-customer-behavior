```python
%%writefile Noaman.py
import streamlit as st
import joblib
import pandas as pd
pipeline_Pre=joblib.load('pipeline_Pre.h5')
pipeline=joblib.load('Loan Prediction ML model.h5')
inputs=joblib.load('input.h5')
ProfessionList=joblib.load('ProfessionList.h5')
CITYList=joblib.load('CITYList.h5')
STATElist=joblib.load('STATElist.h5')

def predict(Income,Age,Experience,Married_Single,House_Ownership,Car_Ownership,Profession,CITY,STATE,CURRENT_JOB_YRS,CURRENT_HOUSE_YRS):
    test_df=pd.DataFrame(columns=inputs)
    test_df.at[0,'Income']=Income
    test_df.at[0,'Age']=Age
    test_df.at[0,'Experience']=Experience
    test_df.at[0,'Married_Single']=Married_Single
    test_df.at[0,'House_Ownership']=House_Ownership
    test_df.at[0,'Car_Ownership']=Car_Ownership
    test_df.at[0,'Profession']=Profession
    test_df.at[0,'CITY']=CITY
    test_df.at[0,'STATE']=STATE
    test_df.at[0,'CURRENT_JOB_YRS']=CURRENT_JOB_YRS
    test_df.at[0,'CURRENT_HOUSE_YRS)']=CURRENT_HOUSE_YRS
    xo= pipeline_Pre.transform(test_df)
    if pipeline.predict(xo)[0] == 0:
        result = "Well Done your Loan Approved"
    elif pipeline.predict(xo)[0] == 1:
        result = "Sorry your loan application rejected"
    return result
def main():
    st.title('Loan Prediction Based on customer behavior')
    Income= st.slider('Income',min_value= 1.0,max_value= 10000000.0,value=10.0,step=500.0)
    Age= st.slider('Age',min_value= 1.0,max_value= 85.0,value=10.0,step=1.0)
    Experience= st.slider('Experience',min_value= 1.0,max_value= 30.0,value=10.0,step=1.0)
    Married_Single= st.selectbox('Married_Single',['single','Married'])
    House_Ownership= st.selectbox('House_Ownership',['rented', 'owned','norent_noown'])
    Car_Ownership=st.selectbox('Car_Ownership',['yes','no'])
    Profession=st.selectbox('Profession',ProfessionList)
    CITY=st.selectbox('CITY',CITYList)
    STATE=st.selectbox('STATE',STATElist)
    CURRENT_JOB_YRS= st.slider('CURRENT_JOB_YRS',min_value= 1.0,max_value= 30.0,value=10.0,step=1.0)
    CURRENT_HOUSE_YRS= st.slider('CURRENT_HOUSE_YRS',min_value= 1.0,max_value= 30.0,value=10.0,step=1.0)
    if st.button('predict'):
        result=predict(Income,Age,Experience,Married_Single,House_Ownership,Car_Ownership,Profession,CITY,STATE,CURRENT_JOB_YRS,CURRENT_HOUSE_YRS)
        st.write(result)
if __name__ =='__main__':
    main()
```
